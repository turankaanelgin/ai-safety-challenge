```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
img=plt.imread('../obstaclemap_fixed.png')
imgray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)*255
imgray[:10,:]=0
imgray[117:,:]=0
imgray[:,:10]=0
imgray[:,115:]=0

ret, thresh = cv.threshold(imgray, 10, 255, 0)
contours, hierarchy = cv.findContours(thresh.astype(np.uint8), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
rects = []
for c in contours:
    rect = cv.boundingRect(c)
    rects.append(rect)
img0 = np.copy(img)
for rect in rects:
    x,y,w,h = rect
    cv2.rectangle(img0,(x,y),(x+w,y+h),(255,0,0),1)
plt.imshow(img0)
```
